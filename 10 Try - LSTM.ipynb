```python
import random
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics 
path = '/project/LSH/'
predict_list_rf = []
acc_list_rf = []
for i in range(10):
    seed_num = 42 + i
    
    random.seed(seed_num)

    data2d = np.load(path + 'x_(7727,4069).npy')
    data2d[data2d>1] = 1
    label = np.load(path + 'y_(7799,1).npy')
    
    idx = list(range(len(data2d)))
    random.shuffle(idx)

    i= round(len(data2d)*0.8)

    X_train = data2d[idx[:i]]
    y_train = label[idx[:i]]

    X_test = data2d[idx[i:]]
    y_test = label[idx[i:]]

    forest = RandomForestClassifier(n_estimators=100)
    forest.fit(X_train,y_train)
    y_pred = forest.predict(X_test)
    acc_list_rf.append(metrics.accuracy_score(y_test, y_pred))
    print(f'정확도 :{metrics.accuracy_score(y_test, y_pred)}, seed_num = {seed_num}')
    predict_list_rf.append(y_pred)
```
