# LABEVENTS

```python
def find_lab_event(event_id):
    abnormal = lab.loc[(lab.ITEMID == event_id) & (lab.FLAG == 'abnormal')]
    print(f"abnormal - min:{abnormal.VALUE.min()}, max:{abnormal.VALUE.max()}")
    normal = lab.loc[(lab.ITEMID == event_id) & (lab.FLAG.isna())]
#     print(f"normal - min:{normal.VALUE.min()}, max:{normal.VALUE.max()}")
    return abnormal, normal
```
